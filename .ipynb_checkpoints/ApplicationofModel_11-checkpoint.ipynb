#### Betting on best EV outcome for each match

```python
def find_highest_ev_bet(df):
    """
    Function to find the highest EV bet for each match.

    Parameters:
    df (DataFrame): A DataFrame containing EV calculations for each outcome.

    Returns:
    DataFrame: The original DataFrame with two new columns:
               'Highest_EV_Bet' indicating the type of bet (Home Win, Draw, Away Win),
               'Highest_EV_Value' indicating the value of the highest EV.
    """
    # Identifying the highest EV bet for each match
    bet_columns = ['Correct_EV_Home_Win', 'Correct_EV_Draw', 'Correct_EV_Away_Win']
    df['Highest_EV_Bet'] = df[bet_columns].idxmax(axis=1)
    df['Highest_EV_Value'] = df[bet_columns].max(axis=1)
    return df

# Applying the function to the DataFrame
EV_calc_df = find_highest_ev_bet(EV_calc_df)
EV_calc_df[['home_team', 'away_team', 'Highest_EV_Bet', 'Highest_EV_Value']].head()
```

```python
# Constants
stake = 10

# Define a function to find the best EV bet for each match
def best_ev_bet(row):
    # Extracting EV values and corresponding odds
    ev_values = {
        'Correct_EV_Home_Win': (row['Correct_EV_Home_Win'], row['Pinnacle Closing Home Win Odds']),
        'Correct_EV_Draw': (row['Correct_EV_Draw'], row['Pinnacle Closing Draw Odds']),
        'Correct_EV_Away_Win': (row['Correct_EV_Away_Win'], row['Pinnacle Closing Away Win Odds'])
    }

    # Filter only positive EV bets
    positive_ev_bets = {k: v for k, v in ev_values.items() if v[0] > 0}

    # If no positive EV bets, return no bet
    if not positive_ev_bets:
        return None, "No Bet"

    # Find the bet with the highest EV
    best_bet = max(positive_ev_bets, key=lambda x: positive_ev_bets[x][0])
    return positive_ev_bets[best_bet][1], best_bet  # Return the odds and type of bet

# Apply the function to find the best bet for each match
EV_calc_df[['Best_Bet_Odds', 'Best_Bet_Type']] = EV_calc_df.apply(best_ev_bet, axis=1, result_type='expand')

# Define a function to calculate the outcome of the best bet
def calculate_best_bet_outcome(row):
    bet_type_to_result_mapping = {
        'Correct_EV_Home_Win': 0,
        'Correct_EV_Draw': 1,
        'Correct_EV_Away_Win': 2
    }
    bet_type = row['Best_Bet_Type']

    if bet_type != "No Bet" and bet_type_to_result_mapping[bet_type] == row['match_result']:
        # Bet won: Calculate profit using fractional odds
        profit = (row['Best_Bet_Odds'] - 1) * stake
        return profit
    elif bet_type != "No Bet":
        # Bet lost
        return -stake
    else:
        # No bet placed
        return 0

# Calculate the outcome of the best bet for each match
EV_calc_df['Best_Bet_Outcome'] = EV_calc_df.apply(calculate_best_bet_outcome, axis=1)

# Calculate the total profit/loss
total_profit_loss_best_bet = EV_calc_df['Best_Bet_Outcome'].sum()

# Displaying total profit/loss and a snippet of the DataFrame
total_profit_loss_best_bet, EV_calc_df[['Best_Bet_Type', 'Best_Bet_Odds', 'Best_Bet_Outcome']].head()

```

```python
# Adjusting the function to bet on all positive EV bets

# Define a function to calculate the outcome of positive EV bets
def calculate_positive_ev_outcomes(row):
    outcomes = {
        'Correct_EV_Home_Win': (0, row['Correct_EV_Home_Win'], row['Pinnacle Closing Home Win Odds']),
        'Correct_EV_Draw': (1, row['Correct_EV_Draw'], row['Pinnacle Closing Draw Odds']),
        'Correct_EV_Away_Win': (2, row['Correct_EV_Away_Win'], row['Pinnacle Closing Away Win Odds'])
    }

    total_outcome = 0
    for bet_type, (result_code, ev, odds) in outcomes.items():
        if ev > 0:
            if result_code == row['match_result']:
                # Bet won
                total_outcome += (odds * stake) - stake
            else:
                # Bet lost
                total_outcome -= stake

    return total_outcome

# Apply the function to calculate outcomes for all positive EV bets
EV_calc_df['Positive_EV_Outcomes'] = EV_calc_df.apply(calculate_positive_ev_outcomes, axis=1)

# Calculate the total profit/loss for all positive EV bets
total_profit_loss_positive_ev = EV_calc_df['Positive_EV_Outcomes'].sum()

total_profit_loss_positive_ev
```

## Kelly Criterion Betting ##

```python
# Define the bankroll and update the no-bet threshold and fraction
initial_bankroll = 1000  # Starting bankroll
bankroll = initial_bankroll
no_bet_threshold = 0.05  # Increase threshold to only place higher confidence bets
fraction = 0.25  # Use a smaller fraction of the Kelly bet

# Function to calculate Kelly bet with bankroll management
def calculate_kelly_bet_with_management(row, bankroll, fraction=1, no_bet_threshold=0.05):
    if bankroll <= 0:
        return 0, "No Bet"  # No bet if bankroll is depleted
    
    # Extract probabilities and corresponding odds
    probabilities = {
        'Home Win': row['Probability_Home_win'],
        'Draw': row['Probability_Draw'],
        'Away Win': row['Probability_Away_win']
    }
    odds = {
        'Home Win': row['Pinnacle Closing Home Win Odds'],
        'Draw': row['Pinnacle Closing Draw Odds'],
        'Away Win': row['Pinnacle Closing Away Win Odds']
    }
    
    # Calculate the Kelly criterion for each bet
    kelly_bets = {}
    for bet_type, p in probabilities.items():
        b = odds[bet_type] - 1
        q = 1 - p
        kelly_fraction = max((b * p - q) / b, 0)  # Kelly criterion formula
        
        # Apply the fraction and check against the no-bet threshold
        kelly_bet_size = kelly_fraction * bankroll * fraction
        if kelly_fraction >= no_bet_threshold:
            kelly_bets[bet_type] = kelly_bet_size
    
    if not kelly_bets:
        return 0, "No Bet"  # No bet if all are below the threshold
    
    # Identify the bet with the highest Kelly fraction
    best_bet_type = max(kelly_bets, key=kelly_bets.get)
    best_bet_amount = kelly_bets[best_bet_type]
    
    # Deduct the bet amount from the bankroll
    bankroll -= best_bet_amount
    
    return best_bet_amount, best_bet_type

# Apply the function to find the Kelly bet for each match
EV_calc_df['Kelly_Bet_Amount'], EV_calc_df['Kelly_Bet_Type'] = zip(*EV_calc_df.apply(calculate_kelly_bet_with_management, axis=1, bankroll=bankroll, fraction=fraction, no_bet_threshold=no_bet_threshold))

# Function to calculate bet outcome and adjust bankroll
def calculate_kelly_bet_outcome_and_adjust(row):
    global bankroll
    bet_type_to_result_mapping = {
        'Home Win': 0,
        'Draw': 1,
        'Away Win': 2
    }
    bet_type = row['Kelly_Bet_Type']

    if bet_type != "No Bet" and bet_type_to_result_mapping[bet_type] == row['match_result']:
        # Bet won: Calculate profit using the bet amount and odds
        profit = (row['Kelly_Bet_Amount'] * (row[f'Pinnacle Closing {bet_type} Odds'])) - row['Kelly_Bet_Amount']
        bankroll += profit  # Add the profit to the bankroll
        return profit
    elif bet_type != "No Bet":
        # Bet lost, bankroll already adjusted when placing the bet
        return -row['Kelly_Bet_Amount']
    else:
        # No bet placed
        return 0

# Apply the function to calculate bet outcomes and adjust the bankroll
EV_calc_df['Kelly_Bet_Outcome'] = EV_calc_df.apply(calculate_kelly_bet_outcome_and_adjust, axis=1)

# Calculate the total profit/loss from Kelly betting with bankroll management
total_profit_loss_kelly = initial_bankroll - bankroll

# Displaying total profit/loss and a snippet of the DataFrame
total_profit_loss_kelly, EV_calc_df[['Kelly_Bet_Type', 'Kelly_Bet_Amount', 'Kelly_Bet_Outcome']].head()
```

```python
import matplotlib.pyplot as plt

# Initialize the bankroll, cumulative profit/loss, and bet counter
initial_bankroll = 1000
bankroll = initial_bankroll
cumulative_profit_loss = []
total_bets = 0

# Define the fraction of the Kelly Criterion to use
kelly_fraction = 0.5  # Adjust this value to change the fraction of Kelly Criterion

# Iterate through each row to calculate cumulative profit/loss
for index, row in EV_calc_df.iterrows():
    if bankroll <= 0:
        cumulative_profit_loss.append(cumulative_profit_loss[-1] if cumulative_profit_loss else 0)
    else:
        # Calculate the bet outcome as before
        bet_type_to_result_mapping = {
            'Home Win': 0,
            'Draw': 1,
            'Away Win': 2
        }
        bet_type = row['Kelly_Bet_Type']
        original_bet_amount = row['Kelly_Bet_Amount']

        # Apply the Kelly fraction to the bet amount
        bet_amount = original_bet_amount * kelly_fraction

        # Check if the EV is greater than or equal to 2 for the selected bet type
        if bet_type == 'Home Win' and row['Correct_EV_Home_Win'] >= 1:
            valid_bet = True
        elif bet_type == 'Draw' and row['Correct_EV_Draw'] >= 1:
            valid_bet = True
        elif bet_type == 'Away Win' and row['Correct_EV_Away_Win'] >= 1:
            valid_bet = True
        else:
            valid_bet = False

        if valid_bet and bet_type != "No Bet":
            total_bets += 1  # Increment the bet counter
            if bet_type_to_result_mapping[bet_type] == row['match_result']:
                # Bet won
                profit = (bet_amount * (row[f'Pinnacle Closing {bet_type} Odds'])) - bet_amount
                bankroll += profit
            else:
                # Bet lost
                profit = -bet_amount
                bankroll += profit
        else:
            profit = 0

        cumulative_profit_loss.append(cumulative_profit_loss[-1] + profit if cumulative_profit_loss else profit)

# Add the cumulative profit/loss to the DataFrame for plotting
EV_calc_df['Cumulative_Kelly_Profit_Loss'] = cumulative_profit_loss

# Plot the cumulative profit/loss
plt.figure(figsize=(12, 6))
plt.plot(EV_calc_df['Cumulative_Kelly_Profit_Loss'], label='Cumulative Profit/Loss', color='b')
plt.axhline(y=0, color='r', linestyle='--', label='Break-Even Line')
plt.title('Cumulative Profit/Loss Over Time Using Kelly Betting Strategy with +1EV Threshold')
plt.xlabel('Number of Bets')
plt.ylabel('Cumulative Profit/Loss')
plt.legend()
plt.grid(True)
plt.show()

# Check final cumulative profit/loss to ensure it matches the total_profit_loss_kelly
final_cumulative_profit_loss = EV_calc_df['Cumulative_Kelly_Profit_Loss'].iloc[-1]

# Print the total number of bets placed
print(f"Total Bets Placed with +2EV Strategy: {total_bets}")
print(f"Final Cumulative Profit/Loss: {final_cumulative_profit_loss:.2f}")
```
